```python
import numpy as np
import pandas as pd
```

```python
log = pd.read_csv("Web_access_log-akumenius.txt",
                   sep = r'\s(?=(?:[^"]*"[^"]*")*[^"]*$)(?![^\[]*\])',
                   engine = 'python',
                   na_values = '-',
                   header = None,
                   usecols=[0, 1, 4, 5, 6, 7, 8, 9], #no column 2 and 3 because there is only '-'
                   names = ['Host','IP','date', 'request', 'status', 'size', 'referer', 'user_agent'],
                   converters= {'size': int, 'status': int})

log.head(10)
```

INFORMATION ON THE VARIABLES

    DNS: DNS (Domain Name System) server. It is a system that helps the user to connect with sites on the Internet by reading and managing the existing domains and identifying the corresponding IP's.
    IP: Client IP address (remote host)
    TIME: The time of entry to the server
    REQUEST: The request log
    STATUS: The status code
    SIZE: The request size
    REFERER: The request header of the HTTP Referer
    USER_AGENT: The request header of the HTTP user agent

```python
from datetime import datetime
from user_agents import parse
from http.client import responses
import numpy as np

logC = log.copy()
```

```python
# Replace "w." and "ww." with "www." in the attribute "host" 
logC["Host"].str.replace("^[w]{1,2}\.","www.")

# We extract the HTTP request methods from the request
logC["request"] = logC["request"].str.extract(r'(GET|HEAD|TRACE|OPTIONS|POST|PUT|DELETE|CONNECT|PATCH)')

# We extract the URI from the request
logC["URI"] = logC["request"].str.extract(r'\s(\/+[^\s]+)')

# We extract the protocol from the request
logC["protocol"] = logC["request"].str.extract(r'(HTTP\/1\.[0|1])')

#status
logC["status"] = logC["status"].apply(lambda x: int(x))

# We maps the status codes to the W3C names.
logC["status_description"] = logC["status"].apply(lambda x: responses[x])

#referer
# Remove quotes
logC["referer"] = logC["referer"].str.replace("\"","")
logC["referer"]= logC["referer"].apply(lambda x: np.nan if x == "-" else x)
logC.head()
```
